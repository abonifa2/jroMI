2-D grid representing antenna area will be plotted. Two modules from covXYs.py will be called to achieve this; geom and covs

geom : sets the magnetic field and antenna geometry

Inputs  : tx,ty,dr
Outputs : px,py,uB,Bo,theta,uphi,uthe

covs : calculates the covariance of voltages for co- and cross-polarized antennas

Inputs  : px,py,uB,Bo,theta,uphi,uthe,Ne,h
Outputs : covXX,covXY

- All the inputs except Ne and h to 'covs' module are generated by the module geom. Hence, Ne and h must be defined explicitly.
- The plots will be drawn for all tx-ty pairs.


```python
import covXYs as covXY
```

```python
geom = covXY.geom
covs = covXY.covs
```

```python
dci = 0.0245    # direction cosine interval

ns  = 6         # number of samples

grid = mgrid[-dci*ns:dci*ns:dci, dci*ns:-dci*ns:-dci] 

tx = grid[0].T[0]; print tx    #  direction cosines specified in a way that loop will begin from the upper left corner
ty = grid[1][0]; print ty
```

```python
scatter(grid[0],grid[1]); xlabel('tx'); ylabel('ty'); scatter(diag(grid[0]),diag(grid[1]),color='r'); # Diagonal shows the N-S line (N is the upper left point)
xlim([-0.16,0.16]); ylim([-0.16,0.16]);
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

dr = 5.
h  = arange(200)*dr
Ne=3.0e12*exp(-h/180.-50.*exp(-h/60.))

gran = tx.shape[0];    # grid sample range 

sf = 0;                # counter of subfigure numbers
fig = figure(figsize = (20,15))

fig.subplots_adjust(hspace=.7,wspace = 0.3)

gs = gridspec.GridSpec(2*ns+1, 2*ns+1)

for ny in range(gran):
    
    for nx in range(gran):

        #ax = fig.add_subplot(gran, gran, sf)
        
        px, py, uB, Bo, theta, uphi, uthe = geom(tx[nx],ty[ny],dr) # since tx=-ty as the the direction cosine sample points on the axes are ordered
        
        covXX, covXY = covs(px, py, uB, Bo, theta, uphi, uthe, Ne, h)
        
        ax = plt.subplot(gs[nx,ny])
        
        ax.plot(h,covXX); ax.plot(h,covXY,color='r')

        ax.set_xticks((0,500,1000)); 
        
        ax.set_yticks((0.000,0.007,0.014));  
        
        sf +=1
        
        if ny != 0:
            
            setp(ax.get_yticklabels(), visible=False)
    
show();
 

        
```
