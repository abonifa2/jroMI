In this ipynb, a solution to improve the performance of covXYs.py will be sought. Since the for loop of the code is the most time-taking part, we will speed this part up. 

This will also be an opportunity to play with h5py 

Here is the method of work:

1) Measure clock time of the current version of the loop.

2) Save the data obtained from "geom" module and make the data ready to be processed by "covs" module.

3) Measure the clock time of the trial version of the loop.



```python
import covXYs as covXY
```

```python
geom = covXY.geom
covs = covXY.covs
```

```python
dci = 0.0245    # direction cosine interval

ns  = 6         # number of samples (default = 6)

grid = mgrid[-dci*ns:dci*ns:dci, dci*ns:-dci*ns:-dci] 

tx = grid[0].T[0]; print tx    #  direction cosines specified in a way that loop will begin from the upper left corner
ty = grid[1][0]; print ty
```

```python
scatter(grid[0],grid[1]); xlabel('tx'); ylabel('ty'); scatter(diag(grid[0]),diag(grid[1]),color='r'); # Diagonal shows the N-S line (N is the upper left point)
xlim([-0.16,0.16]); ylim([-0.16,0.16]);
```

```python
import time

dr = 5.
h  = arange(200)*dr
Ne=8.0e12*exp(-h/180.-50.*exp(-h/60.))

gran = tx.shape[0];    # grid sample range 

sf = 0;                # counter of subfigure numbers
fig = figure(figsize = (20,15))

fig.subplots_adjust(hspace=.7,wspace = 0.3)

stime = time.clock()   # start time

for ny in range(gran):
    
    for nx in range(gran):

        ax = fig.add_subplot(gran, gran, sf)
        
        px, py, uB, Bo, theta, uphi, uthe = geom(tx[nx],ty[ny],dr)
        
        covXX, covXY = covs(px, py, uB, Bo, theta, uphi, uthe, Ne, h)
        
        ax.plot(h,covXX); ax.plot(h,covXY,color='r')

        ax.set_xticks((0,500,1000)); 
        
        ax.set_yticks((0.000,0.015,0.035));  
        
        sf +=1
        
        if nx != 1:
            
            setp(ax.get_yticklabels(), visible=False)
    
etime = time.clock()     # end time

deltime = etime - stime; # time difference

print deltime, "seconds when both (geom and covs) modules are in one loop"

        
```

```python
import h5py as h5
import os
```

It took 39.182063 seconds for the loop to generate the figures by including both modules in the loop

Now, we will save the data that will be input to "covs" module in a .h5 file.

```python
sgtime = time.clock()                     # start time (g for geom)

path = os.getcwd()

f = h5.File(path+'/geom_data.h5','w')     # file is created and ready for further procedures

n = tx.shape[0]*ty.shape[0]               # number of subsets

# group1 will contain the data which don`t change during the propagation throughout all the slabs
group1 = f.create_group("Group1")         
gpx = group1.create_dataset('px_dataset',(n,3), dtype = 'float')
gpy = group1.create_dataset('py_dataset',(n,3), dtype = 'float')

# group200 will contain the local data recalculated at each slab

group200 = f.create_group("Group200")    
gBo = group200.create_dataset('Bo_dataset',[n*200], dtype = 'float')
gtheta = group200.create_dataset('theta_dataset',[n*200], dtype = 'float')
guB = group200.create_dataset('uB_dataset',(n*200,3), dtype = 'float')
guphi = group200.create_dataset('uphi_dataset',(n*200,3), dtype = 'float')
guthe = group200.create_dataset('uthe_dataset',(n*200,3), dtype = 'float')

ns = 0;                          # counter resets

for ny in range(gran):
    
    for nx in range(gran):        
        
        px, py, uB, Bo, theta, uphi, uthe = geom(tx[nx],ty[ny],dr)
    
        gpx.write_direct(px)
        gpy.write_direct(py)

        start = ns*200
        stop  = (ns+1)*200
        
        gBo[start:stop] = Bo
        gtheta[start:stop] = theta
        guB[start:stop] = uB
        guphi[start:stop] = uphi
        guthe[start:stop] = uthe

        ns+=1

egtime = time.clock()     # end time

delgtime = egtime - sgtime; # time difference

print delgtime, "seconds to generate and save the data"
```

As seen above, most of the total time is consumed on generating and saving the data needed for "covs" module. This group of data is used only one. Hence, it is a better idea to process it in a seperate loop and save in a file for future purposes.

Now, we will build a similar loop for "covs" module.

```python
rpx = fr['/Group1/px_dataset']         # r for read
rpy = fr['/Group1/py_dataset']

rBo = fr['/Group200/Bo_dataset']
rtheta = fr['/Group200/theta_dataset']
ruB = fr['/Group200/uB_dataset']
ruphi = fr['/Group200/uphi_dataset']
ruthe = fr['/Group200/uthe_dataset']
```

```python
sf = 0;                # counter of subfigure numbers

fig = figure(figsize = (20,15))

fig.subplots_adjust(hspace=.7,wspace = 0.3)

sctime = time.clock()      # s for start, c for covs

nc = 0;

for ny in range(gran):
    
    for nx in range(gran):
        
        ax = fig.add_subplot(gran, gran, sf)
       
        start = nc*200; st = start
        stop  = (nc+1)*200; sp = stop

        covXX, covXY = covs(rpx[nc], rpy[nc], ruB[st:sp], rBo[st:sp], rtheta[st:sp], ruphi[st:sp], ruthe[st:sp], Ne, h)
        
        nc +=1;
        
        ax.plot(h,covXX); ax.plot(h,covXY,color='r')

        ax.set_xticks((0,500,1000)); 
        
        ax.set_yticks((0.000,0.015,0.035));  
        
        sf +=1
        
        if nx != 1:
            
            setp(ax.get_yticklabels(), visible=False)

ectime = time.clock()

delctime = ectime - sctime;

print delctime, "seconds when data is loaded from a .h5 file explicitly"
```
