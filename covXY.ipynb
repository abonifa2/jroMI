```python
import jropack.igrf.igrf11 as igrf11 #imports the entire *namespace* of igrf11
igrf_B=igrf11.igrf_B                 # IGRF subroutine that returns geomagetic field vector Bv
a_igrf=igrf11.a                      # [km] mean Earth radius used in IGRF model
a_igrf
```

```python
import scipy.constants as pc     #physical constants
qe=pc.e                          # [C] electric charge
me=pc.m_e                        # [kg] mass of an electron
eps0=pc.epsilon_0                # [F/m] free space electrical permittivity
mu0=pc.mu_0                      # [H/m] free space magnetic permeability
c=pc.c                           # [m/s] light speed

print qe,me,eps0,mu0,c

#wgs84 model of spheroid Earth
a=6378.137                                # [km] - equatorial radius WGS 84
flatness=1/298.257223563                         # flattening f = (a-b)/a
b=a*(1-flatness)                                 # [km] - polar radius
eccentricity=sqrt(a**2-b**2)/a            # eccentricity

#spacetime of JRO
year=2004.99     # experiment time needed in IGRF model
latR=-11.947917*pi/180.; lonR=-76.872306*pi/180.  #geodedic --- like GPS indicates
hR =0.463                                  # [km] - local height above reference ellipsoid
nR=a/sqrt(1-(flatness*(2-flatness)*sin(latR)**2))        # radius of curvature
xR=(nR+hR)*cos(latR)*cos(lonR)
yR=(nR+hR)*cos(latR)*sin(lonR)
zR=(nR*(1-eccentricity**2)+hR)*sin(latR)

#north, east and zenith unit vectors at JRO
rR=sqrt(xR**2+yR**2+zR**2)
pR=sqrt(xR**2+yR**2)
uRr=array([xR,yR,zR])/rR                                          # Radial unit vector
uRe=array([-yR,xR,0])/pR                                          # East  unit vector
uRz=array([cos(latR)*cos(lonR),cos(latR)*sin(lonR),sin(latR)])    # Zenith unit vector
uRn=cross(uRz,uRe)                                                # North unit vector

#orthonormal basis vectors including the JRO on-axis direction
dec=-12.88*pi/180.; ha=-(4.+37./60.)*pi/180.                         
uo=array([cos(dec)*cos(ha/4.+lonR),cos(dec)*sin(ha/4.+lonR),sin(dec)])    # on axis
ux=cross(uRz,uo); ux=ux/norm(ux)                                        # along building to the right
uy=cross(uo,ux) 

print dot(uo,uo)
ux,uy,uo
```

```python
#angular position of target in terms of direction cosines tx and ty
tx=-1*0.0248; ty=-tx;  #perp to B with -1
tz=sqrt(1.-(tx**2.+ty**2.))
uk=tx*ux+ty*uy+tz*uo  # propagation direction to the target 
#polarization vectors at ground
px=cross(uk,cross(uk,ux))
px=px/norm(px)
py=cross(uk,cross(uk,uy))
py=py/norm(py)

xyzT=array([xR,yR,zR])                            #initialize target vector
uBs=[];Bos=[];thetas=[];uphis=[];uthes=[]; #initialize magneto-ionic parameter lists
dr=5.                                      # [km] range step
for i in range(200):                       #loop to targets at different ranges 0, dr, 2*dr, ... 199*dr
    rT=norm(xyzT)
    pT=sqrt(xyzT[0]**2+xyzT[1]**2)
    latT=arctan2(xyzT[2],pT)
    lonT=arctan2(xyzT[1],xyzT[0]) 
    urT=xyzT/rT                          #radial unit vector at target location
    ueT=array([-xyzT[1],xyzT[0],0])/pT   #east unit vector = -x*sin(lonR) + y*cos(lonR)
    unT=-cross(ueT,urT)                  #north unit vector   
    unT=unT/norm(unT)
    
    XB,YB,ZB,B=igrf_B(year,rT-a_igrf,rad2deg(lonT),rad2deg(latT))   # X = North, Y = East, Z = Down (nT)
    Bv=XB*unT+YB*ueT-ZB*urT     # B-field vector
    Bo=norm(Bv)                 # [nT] where it is necessary multiply it by 10^-09 (1 nT = 10^-05 Gauss and 1 Gauss = 10^-04 kg/(C.s)
    uB=Bv/Bo                    # unit vector for magnetic field

    theta=arccos(dot(uB,uk))   # [rad] the angle theta between uB and uk            
    uphi=cross(uB,uk)          # phi unit vector
    uphi=uphi/norm(uphi)
    uthe=cross(uphi,uk)        # theta unit vector
    uthe=uthe/norm(uthe)

    uBs=uBs+[uB]; Bos=Bos+[Bo]; thetas=thetas+[theta]; uphis=uphis+[uphi]; uthes=uthes+[uthe] #add to lists
    xyzT=xyzT+dr*uk #increment the target position vector

uB=array(uBs); Bo=array(Bos); theta=array(thetas); uphi=array(uphis); uthe=array(uthes) #convert lists to arrays

#radar constants
f0=49.98e06      # [Hz] operating frequency of radar
w0=2*pi*f0       # [rad/s] angular operating frequency of radar
k0=w0/c          # free space propagation constant

#ionosphere specification
dr=5.                                           # [km] slab width
h=arange(200)*dr                                # [km] height grid vector
Ne=8.0e12*exp(-h/180.-50.*exp(-h/60.)) 
figure(figsize=(15,9))
subplot(3,4,1); plot(h,Ne)

#magneto-ionic parameters at target ranges (Appleton-Hartree eqn)
X=(Ne*qe**2)/(eps0*me)/w0**2                       # normalized plasma frequency
Y=qe*Bo*1.0e-09/me/w0                              # normalized gyro frequency
YL=Y*cos(theta); YT=Y*sin(theta) 
F1=YT**2/(1-X)/2.; F2=sqrt(F1**2+YL**2)
FX=F1+F2; FO=F1-F2
no=sqrt(1-(X/(1-FO))); nx=sqrt(1-(X/(1-FX)))
deln=(no-nx)/2.                                 # difference between the refractive indices
avn=(no+nx)/2.                                  # avg refractive indices
ax=FO/YL                                        # axial ratio showing the polarization changes

denom=1.+ax**2
dphase=k0*deln*dr*1000.
eterm=exp(1j*dphase)
AA=(ax**2*eterm+conj(eterm))/denom
BB=2*ax*sin(dphase)/denom

subplot(3,4,2);plot(h,theta*180./pi-90)
subplot(3,4,3);plot(h,X)
subplot(3,4,4);plot(h,Y,'.');plot(h,YL,'r');plot(h,YT,'b')
subplot(3,4,5);plot(h,no),plot(h,nx)
subplot(3,4,6);plot(h,ax)
subplot(3,4,7);plot(h,denom)
subplot(3,4,8);plot(h,dphase)
subplot(3,4,9);plot(h,real(AA))
subplot(3,4,10);plot(h,imag(AA))
subplot(3,4,11);plot(h,real(BB))

vxx=[];vxy=[]
Tx=mat(px).T               #Tx'ed unit vector
Rx=mat(px)                 #Rx'ing unit vector  
Ry=mat(py)
for i in range(200):    
    M=mat([[AA[i],BB[i]],[-BB[i],conj(AA[i])]]) #propagator of vector [Eth;Eph] across slab i
    S=mat([uthe[i],uphi[i]])                    #transforms field vector from JROxyo- to th/phi-system
    P=S.T*(M*S)                                 #field propagator across slab i in xyo-system        
    Tx=P*Tx                                     #upgoing polarization vector above slab i
    Rx=Rx*P                                     #"back propagates" sampling unit vector to above slab i
    Ry=Ry*P
    vxx=vxx+[Rx*Tx]                             #scattered x-pol voltage from the top of slab i 
    vxy=vxy+[Ry*Tx]                             #scattered y-pol voltage from the top of slab i 
vxx=array(vxx);vxy=array(vxy)
subplot(3,4,12);plot(h,X*abs(vxx[:,0,0])**2); xlabel('Height (km)');plot(h,X*abs(vxy[:,0,0])**2);grid('on')
```
