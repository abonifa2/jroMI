```python
import Grid2D as G2D

Gr2D = G2D.grid_2D
FigGrid = G2D.FigGrid
CovFigGrid = G2D.CovFigGrid
```

```python
dci = 0.0245
ns = 6
grid,tx,ty = Gr2D(0.0245,6)
```

```python
dr = 5.
h  = arange(200)*dr
Ne=8.0e12*exp(-h/180.-50.*exp(-h/60.))

FigGrid(grid,dci,ns)
```

```python
CovFigGrid(tx,ty,dr,Ne,h)
```
