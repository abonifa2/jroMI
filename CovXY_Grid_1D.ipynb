```python
import covXYs as covXY
```

```python
geom = covXY.geom
covs = covXY.covs
```

```python
dr = 5.
h = arange(200)*dr
```

```python
fig = figure(figsize = (12,10))

fig.subplots_adjust(hspace=1,wspace = 0.5)

for ir in range(12):
    tx = -0.0245*(ir-4); ty = -tx
    print ir, tx, ty
    px, py, uB, Bo, theta, uphi, uthe = geom(tx,ty,dr)
    Ne=8.0e12*exp(-h/180.-50.*exp(-h/60.))     
    covXX, covXY = covs(px,py,uB,Bo,theta,uphi,uthe,Ne,h)
    ax = fig.add_subplot(12, 1, ir+1)
        
    ax.plot(h,covXX); # ax.set_xlabel('Height (km)'); 
    ax.plot(h,covXY,color='r')
    
    ax.set_xticks((0,250,500,750,1000)); 
        
    ax.set_yticks((0.000,0.015,0.035));  
             
    if ir != 11:
        
        setp(ax.get_xticklabels(), visible=False)
    
    figure(2)
    scatter(tx,ty); xlabel('tx'); ylabel('ty'); title('Scanning in N-S directions')    
```
